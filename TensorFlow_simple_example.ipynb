Import packages and data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Churn.csv")
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))
y = df['Churn'].apply(lambda x : 1 if x == 'Yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.2)
X_train.head()
y_train.head()
```

Import dependencies 

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

Build the model 

```python
#A Sequential model is appropriate for a plain stack of 
# layers where each layer has exactly one input tensor and one output tensor.

model = Sequential() 
#add layers to the neural network
#units represent the number of the neurons

#input dimension represents the number of the input which is the number of the columns 
#in the train set (x)
#so these following are three hidden layer
model.add(Dense(units = 32, activation = 'relu', input_dim=len(X_train.columns)))
model.add(Dense(units = 64, activation='relu'))
model.add(Dense(units=1, activation= 'sigmoid'))

#Important: the number of unit in your last layer will predict 
#what your output will look alike. In this case we want to predict "y" or "n" so we have 1 neuron
```

Compile the model: This means that we can specify how we want to train our model 

```python
#The purpose of loss functions is to compute 
# the quantity that a model should seek to minimize during training.
model.compile(loss = 'binary_crossentropy', optimizer = 'sgd', metrics = 'accuracy')
```

Fit, predict and evaluate

```python
#this is the model training
model.fit(X_train, y_train, epochs = 200, batch_size = 32)

#in general what we want to see is: increasing the accuracy and decreasing the loss 
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
accuracy_score(y_test, y_hat)
```

Saving and reloading

```python
model.save('tfmodel')

#del model if you want to delete the mode
model = load_model('tfmodel')
```
